# defenseCounts

```python
#comboCounts data analysis
import pandas as pd


defenseSuccess=[]
musicScene__=[]
defenseFail=[]
country_oo=[]

for i in range(len(record)):
    if record[i]['name']=='defenseCounts':
        custom_parameters_oo=record[i]['custom_params']
        defenseSuccess.append(list(custom_parameters_oo.items())[0][1])
        musicScene__.append(list(custom_parameters_oo.items())[1][1])
        defenseFail.append(list(custom_parameters_oo.items())[2][1])
        country_oo.append(record[i]['country'])

zipped = list(zip(defenseSuccess,musicScene__,defenseFail,country_oo))
df_defenseCounts = pd.DataFrame(zipped, columns=['Defense Success', 'Music Scene','Defense Fail','Country'])
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):

    
#     print(df_defenseCounts)   
    dataframe_=df_defenseCounts.sort_values(by='Defense Success', ascending=True,ignore_index=True)
#     print(dataframe_)
    b=dataframe_.iloc[26:80]
#     print(b)
    b.to_csv('defenseCounts_results.csv', index=False)
    
    
```
