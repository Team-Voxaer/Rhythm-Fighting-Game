```python
#comboCounts data analysis
import pandas as pd
import json
record = []
for line in open('0701', 'r'):
    record.append(json.loads(line))
    
sword=[]
healing=[]
defend=[]
attack=[]
musicScene_o=[]
grandCross=[]
thunder=[]
country_o=[]

for i in range(len(record)):
    if record[i]['name']=='comboCounts':
        custom_parameters_o=record[i]['custom_params']
        sword.append(list(custom_parameters_o.items())[0][1])
        healing.append(list(custom_parameters_o.items())[1][1])
        defend.append(list(custom_parameters_o.items())[2][1])
        attack.append(list(custom_parameters_o.items())[3][1])
        musicScene_o.append(list(custom_parameters_o.items())[4][1])
        grandCross.append(list(custom_parameters_o.items())[5][1])
        thunder.append(list(custom_parameters_o.items())[6][1])
        country_o.append(record[i]['country'])

zipped = list(zip(sword,healing,defend,attack,musicScene_o,grandCross,thunder,country_o))
df_comboCounts = pd.DataFrame(zipped, columns=['Sword', 'Healing','Defend','Attack','Music Scene','Grand Cross',
                                               'Thunder','Country'])

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):

    df_comboCounts.to_csv('comboCounts_results.csv', index=False)
    
    
```

```python
import pandas as pd
from matplotlib import pyplot as plt
 
data = {'All The Way North Easy.mid':5, 'All The Way North Normal.mid':9,
        'All_The_Way_North_Hard.mid':23,'Never Gonna Give You Up Easy.mid':20,
        'Never Gonna Give You Up Normal.mid':10,
        'Never_Gonna_Give_You_Up_Hard.mid':10}
music_name = list(data.keys())
values = list(data.values())
  
 
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(music_name, values)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 15, fontweight ='bold',
             color ='black')
 
# add title
ax.set_title("Chosen Musics",
             loc ='left', fontsize = 15, fontweight ='bold')
 
# Add Text watermark
fig.text(0.9, 0.15, "No. of Occurences", fontsize = 15,
         color ='black', ha ='right', va ='bottom', fontweight ='bold' )
 
    
# Show Plot
plt.show()
```
